```python
import pandas as pd

df['Age Group'] = 'Other'  


df.loc[(df['Age'] >= 18) & (df['Age'] <= 25), 'Age Group'] = '18-25'
df.loc[(df['Age'] >= 26) & (df['Age'] <= 35), 'Age Group'] = '26-35'
df.loc[(df['Age'] >= 36) & (df['Age'] <= 45), 'Age Group'] = '36-45'
df.loc[(df['Age'] >= 46) & (df['Age'] <= 55), 'Age Group'] = '46-55'
df.loc[(df['Age'] >= 56) & (df['Age'] <= 65), 'Age Group'] = '56-65'
df.loc[df['Age'] >= 66, 'Age Group'] = '65+'

print(df)
```

```python
age_group = df['Age Group'].value_counts()
age_group.reset_index()
```

```python
df.loc[(df['Size'] == 'L') , 'Size'] = 'Large'
df.loc[(df['Size'] == 'S') , 'Size'] = 'Small'
df.loc[(df['Size'] == 'M') , 'Size'] = 'Meduim'
df.loc[(df['Size'] == 'XL') , 'Size'] = 'XLarege'


df
```

```python
# What is the average age of customers?

average_age = (df['Age'].mean()).round(2)
average_age
```

```python
# What is the gender distribution of customers?

df['Gender'].value_counts().reset_index()
```

```python
# What The sum amount Customer Age group ?

df.groupby(['Age Group'])['Purchase Amount (USD)'].sum().reset_index()
```

```python
# What The Average Amount For Each Gender?

(df.groupby(['Gender'])['Purchase Amount (USD)'].mean().reset_index()).round(2)
```

Frequency of Purchases

```python
# What are the most common frequencies of purchases?

df['Frequency of Purchases'].value_counts().reset_index()
```

```python
# What Age Has The Highest Frequencies?

Frequency_of_Purchases = df.groupby(['Age Group'])['Frequency of Purchases'].count().reset_index()
Frequency_of_Purchases.sort_values(by= 'Frequency of Purchases' , ascending= False)
```

```python
# What Category has the highest frequencies?

Frequency_of_Purchase = df.groupby(['Category'])['Frequency of Purchases'].count().reset_index()
Frequency_of_Purchase.sort_values(by= 'Frequency of Purchases' , ascending= False)
```
